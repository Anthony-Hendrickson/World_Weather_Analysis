```python
# np.random.uniform to generate an array of 2,000 numbers for latitude and longitudes
lat = np.random.uniform(low=-90.000, high=90.000, size=2000)
lng = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lng = zip(lat, lng)
coordinates = list(lat_lng)
city_coordinates_df = pd.DataFrame(coordinates)
city_coordinates_df.head(10)
```

```python
# rename columns
city_coordinates_df.rename(columns = {0:'lat',1:'lng'},inplace = True)
city_coordinates_df["city_name"] = ""
city_coordinates_df["country_code"] = ""
city_coordinates_df
```

```python
# Create a list for holding the cities.
cities = []
countries = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    country = citipy.nearest_city(coordinate[0], coordinate[1]).country_code
    cities.append(city)
    countries.append(countries)
# Print the city count to confirm sufficient count.
len(cities)
```

```python
city_coordinates_df['city'] = cities
city_coordinates_df['country'] = countries
```
